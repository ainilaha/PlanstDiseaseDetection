## Splite dataset

save 10% of each class as validation data set

```python
import random
import os
import shutil
def split_train_test_data(path,val_path,percentage=10):
    for subpath in os.listdir(path):
        imgs_list = os.listdir(os.path.join(path,subpath))
        k = len(imgs_list) * percentage // 100
        indicies = random.sample(range(len(imgs_list)), k)
        if not os.path.isdir(os.path.join(val_path,subpath)):
            os.mkdir(os.path.join(val_path,subpath))
        for i in indicies:
            shutil.move(os.path.join(path,subpath,imgs_list[i]),
                        os.path.join(val_path,subpath,imgs_list[i])                              )
            print("Moving:",imgs_list[i])
```

```python
path = "D:/Projects/crowdai_train/train"
val_path = "D:/Projects/crowdai_train/val"
split_train_test_data(path,val_path)
```
